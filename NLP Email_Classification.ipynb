```python
# importing libraries
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
# loading and making copy of data for safe side
data = pd.read_csv(r"C:\Users\Prem\Desktop\ML projects\spam.csv")
df = data
```

```python
#  text cleaning 
stop_words = set(stopwords.words('english'))  ##stopwords only supports certain predefined methods, such as: -.words()

def clean_text(text):
    tokens = word_tokenize(text.lower())
    words = [ w for w in tokens if w.isalpha() and w not in stop_words]
    return ' '.join(words)

df['Message'] = df['Message'].apply(clean_text)
```

```python
# splitting features and label
x = df['Message']
y = df['Category']
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
# vectorization
vectorizer = TfidfVectorizer(ngram_range=(2,3))
x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
```

```python
# model buildng 
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
# predict the model
y_pred = model.predict(x_test)
```

```python
# evaluation 
acc = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# print scores
print('accuracy_score:',acc)
print('\nclassification_report:',report)
```

```python
# prediction function
def predict_email(text):
    text = clean_text(text)
    vectors = vectorizer.transform([text])
    prediction = model.predict(vectors)
    return prediction[0]
```

```python
# testing it 
print(predict_email("Hey, are we meeting tomorrow?"))
print(predict_email("congratulations you won free cash prize"))
```
